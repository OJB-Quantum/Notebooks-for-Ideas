```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert
```

```python
# Generate a wave packet with a Gaussian envelope
fs = 1000  # Sampling frequency in Hz
t = np.arange(0, 1, 1/fs)  # Time vector
frequency = 5  # Frequency of the sine wave
amplitude = 1  # Amplitude of the sine wave
wave_packet = amplitude * np.sin(2 * np.pi * frequency * t) * np.exp(-(t-0.5)**2 / 0.01)

# Compute the envelope using the Hilbert transform
analytical_signal = hilbert(wave_packet)
envelope = np.abs(analytical_signal)

# Plotting
plt.figure(figsize=(10, 6), dpi=400)
plt.plot(t, wave_packet, label='Wave Packet', color='green')
plt.plot(t, envelope, label='Envelope', linestyle='--', color='blue')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend()
plt.title('Wave Packet and Its Envelope')
plt.show()
```
