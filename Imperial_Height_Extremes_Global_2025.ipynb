```python
import matplotlib.pyplot as plt
import pandas as pd

# Use rcParams to set the default color cycle and DPI
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=['blue'])
plt.rcParams['figure.dpi'] = 250


# Data for the plot
heights = [
    "7 ft 2 in", "7 ft 3 in", "7 ft 4 in", "7 ft 5 in", "7 ft 6 in",
    "7 ft 7 in", "7 ft 8 in", "7 ft 9 in", "7 ft 10 in", "7 ft 11 in",
    "8 ft or higher"
]
estimated_numbers = [510, 300, 180, 90, 55, 35, 18, 12, 6, 4, 3]

# Create a pandas DataFrame
df = pd.DataFrame({'Height': heights, 'Estimated Number of People': estimated_numbers})

# Sort the DataFrame to have the bars in descending order for plotting
df = df.sort_values('Estimated Number of People', ascending=False)

# Create the plot
plt.figure(figsize=(12, 8))

# Plot the bars
bars = plt.bar(df['Height'], df['Estimated Number of People'])

# --- Code to add labels on top of each bar ---
for bar in bars:
    yval = bar.get_height()
    # Place text slightly above the bar, centered horizontally
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 5, int(yval), ha='center', va='bottom')
# ---------------------------------------------

# Add labels and title for clarity
plt.xlabel('Height')
plt.ylabel('Estimated Number of Living People')
plt.title('Estimated Number of Living People in the World by Height')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add a bit of padding to the y-axis to make room for the top label
plt.ylim(0, max(estimated_numbers) * 1.1)

# Ensure everything fits nicely
#plt.tight_layout()

# Display the plot
#plt.show()
```
