## Gabriel's Horn

$
V=\pi \int_1^a\left(\frac{1}{x}\right)^2 \mathrm{~d} x=\pi\left(1-\frac{1}{a}\right) \quad \lim _{a \rightarrow \infty} V=\lim _{a \rightarrow \infty} \pi\left(1-\frac{1}{a}\right)=\pi
$

$
A=2 \pi \int_1^a \frac{1}{x} \sqrt{1+\left(-\frac{1}{x^2}\right)^2} \mathrm{~d} x>2 \pi \int_1^a \frac{\mathrm{~d} x}{x}=2 \pi \ln (a) \quad \lim _{a \rightarrow \infty} A \geq \lim _{a \rightarrow \infty} 2 \pi \ln (a)=\infty
$


```python
# Gabriel's Horn, without grid lines

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Onri prefers using rcParams to increase the quality of all plots to be higher than the default settings.
# This only needs to be done once at beginning of every Jupyter Notebook.
plt.rcParams['figure.dpi'] = 200

def gabriels_horn():
    # Define x values from 1 to 10 (finite truncation for visualization)
    x = np.linspace(1, 10, 300)
    y = 1 / x  # Function y = 1/x

    # Create theta values for rotation
    theta = np.linspace(0, 2*np.pi, 100)

    # Create a meshgrid for revolution
    X, T = np.meshgrid(x, theta)
    Y = 1 / X
    Z = Y * np.sin(T)
    Y = Y * np.cos(T)

    # Plot the surface
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, color='b', alpha=0.6, edgecolor='none')

    # Remove grid lines
    ax.grid(False)

    # Labels and title
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")
    ax.set_zlabel("Z-axis")
    ax.set_title("Gabriel's Horn")

    # Display the plot
    plt.show()

# Run visualization
gabriels_horn()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def volume_vs_surface_area():
    a_values = np.linspace(1, 100, 1000)  # Range for 'a'
    volume = np.pi * (1 - 1 / a_values)
    surface_area = 2 * np.pi * np.log(a_values)

    # Plot
    plt.figure(figsize=(8, 6))
    plt.plot(a_values, volume, label="Volume $V = \pi(1 - 1/a)$", color='b')
    plt.plot(a_values, surface_area, label="Surface Area $A \geq 2\pi \ln(a)$", color='r')

    plt.axhline(y=np.pi, color='b', linestyle='dashed', label="$\pi$ Limit")
    plt.axhline(y=0, color='black', linewidth=0.5)

    plt.xlabel("a")
    plt.ylabel("Value")
    plt.title("Volume vs Surface Area of Gabriel's Horn")
    plt.legend()
    plt.grid(True, linestyle='--', linewidth=0.5)

    # Show plot
    plt.show()

# Run visualization
volume_vs_surface_area()
```
