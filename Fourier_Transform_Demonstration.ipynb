```
# -*- coding: utf-8 -*-
"""
## Fourier Transform: Before and After

This Google Colab notebook demonstrates the effect of applying a Fourier Transform to a signal.

1.  **"Before" - The Time Domain:** We create a signal by combining several sine waves of different frequencies.
2.  **"After" - The Frequency Domain:** We then apply the Fast Fourier Transform (FFT) to see the signal's constituent frequencies.
"""

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Generate the Signal (Time Domain) ---

# Parameters for the signal
sampling_rate = 1000  # Number of samples per second (Hz)
t = np.linspace(0, 1, sampling_rate, endpoint=False)  # Time vector for 1 second

# Create a composite signal by adding three sine waves
# - A 5 Hz sine wave with amplitude 3
# - A 20 Hz sine wave with amplitude 1
# - A 50 Hz sine wave with amplitude 2
signal = 3 * np.sin(2 * np.pi * 5 * t) + 1 * np.sin(2 * np.pi * 20 * t) + 2 * np.sin(2 * np.pi * 50 * t)

# --- 2. Plot the Original Signal (Before) ---

plt.figure(figsize=(14, 6), dpi=250)
plt.subplot(1, 2, 1)  # Create a subplot for the time-domain signal
# Force the plot color to be pure blue
plt.plot(t, signal, color='blue')
plt.title("Before: Signal in Time Domain")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.xlim(0, 0.5) # Zoom in to see the wave shape more clearly

# --- 3. Apply Fourier Transform (FFT) ---

# Compute the Fast Fourier Transform
fft_result = np.fft.fft(signal)

# Compute the frequency bins for the x-axis
# N is the number of sample points
N = len(signal)
frequencies = np.fft.fftfreq(N, 1/sampling_rate)

# Calculate the absolute value of the FFT result to get the amplitude
# We only need to plot the first half of the frequencies (positive frequencies)
# We also normalize the amplitude by N/2
amplitude = 2.0/N * np.abs(fft_result[:N//2])
positive_frequencies = frequencies[:N//2]


# --- 4. Plot the Transformed Signal (After) ---

plt.subplot(1, 2, 2) # Create a subplot for the frequency-domain signal
# Force the bar color to be pure blue
plt.bar(positive_frequencies, amplitude, color='#0000FF')
plt.title("After: Signal in Frequency Domain")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.xlim(0, 100) # Zoom in on the relevant frequency range

# Display the plots
plt.tight_layout() # Adjust plots to prevent overlap
plt.show()

"""
### Interpretation of the Plots

* **Before (Left Plot):** The first plot shows our signal plotted against time. The `color='#0000FF'` argument in the `plt.plot()` function call forces this line to be pure blue.

* **After (Right Plot):** The second plot shows the frequency-domain representation. Similarly, the `color` argument in the `plt.bar()` call makes the bars pure blue. This method gives you precise control over each plot's appearance, independent of global settings.
"""
```
