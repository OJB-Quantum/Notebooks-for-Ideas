```python
import skrf as rf
import matplotlib.pyplot as plt
```

```python
# Create a frequency object for a single frequency (e.g., 1 GHz)
freq = rf.Frequency(1, 1, npoints=1, unit='GHz')

# Create a Network for the load impedance (e.g., 100 + j50 Ohms)
z_load = 100 + 50j
load_network = rf.Network(frequency=freq, z0=50)  # 50 Ohms system impedance
load_network.z = [[z_load]]  # Setting the load impedance

# Plotting the initial impedance on the Smith chart
fig, ax = plt.subplots(dpi=300)
rf.plotting.plot_smith(load_network, chart_type='z', ax=ax, title='Impedance Matching on Smith Chart')

# For demonstration, let's assume we want to move towards matched impedance
# We'll just mark the center (matched condition) for visual reference
ax.plot(0, 0, 'ro')  # Matched condition at the center

plt.show()
```
