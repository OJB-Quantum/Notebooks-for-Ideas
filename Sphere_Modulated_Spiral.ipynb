```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters for the hemisphere modulated spiral
num_turns = 5  # Number of turns in the spiral
num_points = 1000  # Number of points in the spiral
radius = 5  # Base radius of the spiral

# Generate theta values for the full sphere spiral
theta = np.linspace(0, num_turns * 2 * np.pi, num_points)

# Modulation function for the hemisphere effect
modulation = np.sqrt(np.maximum(0, 1 - (theta / (num_turns * 2 * np.pi))**2))

# Compute x, y, and z coordinates for the upper hemisphere
x_upper = radius * np.cos(theta) * modulation
y_upper = radius * np.sin(theta) * modulation
z_upper = np.linspace(0, radius, num_points)

# Compute x, y, and z coordinates for the lower hemisphere
x_lower = x_upper
y_lower = y_upper
z_lower = -z_upper  # Reflect across the origin

# Plotting the 3D full sphere modulated spiral
fig = plt.figure(figsize=(8, 8), dpi=200)
ax = fig.add_subplot(111, projection='3d')

# Plot both hemispheres
ax.plot(x_upper, y_upper, z_upper, lw=2, color='b')
ax.plot(x_lower, y_lower, z_lower, lw=2, color='r')

# Turn off the grid
ax.grid(False)

ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("Full Sphere Modulated Spiral in 3D")

plt.show()
```
