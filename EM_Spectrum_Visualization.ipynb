Sine wave increasing in frequency to visuelize the EM spectrum for presentation & publication figures

```
import numpy as np
import matplotlib.pyplot as plt

# --- Control Knob ---
amplitude = 1.0         # Change the amplitude of the wave
start_frequency = 0.005     # Starting frequency in Hz
end_frequency = 2.5      # Ending frequency in Hz
# --------------------

# Set the DPI using rcParams
plt.rcParams['figure.dpi'] = 300

# Set the figure size to 7 inches wide by 2 inches tall
plt.figure(figsize=(7,1.5))

# Generate the data
# Create a time vector
t = np.linspace(0, 10, 10000)

# Create a frequency vector that linearly increases based on the control knob settings
frequency = np.linspace(start_frequency, end_frequency, 10000)

# Generate the sine wave with the increasing frequency and specified amplitude
sine_wave = amplitude * np.sin(2 * np.pi * frequency * t)

# Create the plot with increased line thickness
plt.plot(t, sine_wave, color='blue', linewidth=1.5)

# Add labels and title for clarity
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Sine Wave with Increasing Frequency")

# Use tight_layout to ensure labels fit within the figure
plt.tight_layout()
```
