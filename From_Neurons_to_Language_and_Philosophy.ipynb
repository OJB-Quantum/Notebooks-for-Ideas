```python
# Install Graphviz (system) and the Python wrapper
#     • Installs ~50 MB, takes <1 min on Colab
#
#@title Install Graphviz ↴ {display-mode: "form"}
%%bash
apt-get -qq update
apt-get -qq install -y graphviz
pip install --quiet graphviz
echo "Graphviz & python-graphviz installed"
```

```python
#  Define the DOT code for the mind map
#
#@title Define DOT string ↴ {display-mode: "form"}
dot_source = r"""
digraph MindMap {
    rankdir=LR;                       // left-to-right layout
    node [shape=box, style="rounded,filled", fillcolor="#e0f2ff"];

    "Neurons" -> "Synaptic\nconnections" -> "Neural circuits\n& networks" ->
    "Core affect\n& drives" -> "Curiosity" -> "Metacognition\n& language" ->
    "Philosophy";

    "Philosophy" -> "Specialised\nsub-fields";
    "Philosophy" -> "Institutional\nembedding";
    "Philosophy" -> "Downstream\ninfluence";
}
"""
print(dot_source)  # sanity-check (optional)
```

```python
#  Render the graph to PNG & SVG, then display PNG inline
#
#@title Render and display ↴ {display-mode: "form"}
from graphviz import Source
from IPython.display import Image, SVG, display

# Render to PNG for quick preview
Source(dot_source, filename="mindmap", format="png").render(cleanup=True)
display(Image("mindmap.png"))

# Also save an SVG (vector) copy for high-quality export
Source(dot_source, filename="mindmap", format="svg").render(cleanup=True)
print("Files saved: mindmap.png  |  mindmap.svg")
```

```python
# SVG is crisp at any zoom level
#
#@title Display SVG inline ↴ {display-mode: "form"}
display(SVG("mindmap.svg"))
```
