```python
import matplotlib.pyplot as plt
import numpy as np

# This only needs to be done once at beginning of every Jupyter Notebook.
plt.rcParams['figure.dpi'] = 200

# High contrast plotting style via rcParams
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['figure.facecolor'] = '#000000'
plt.rcParams['axes.facecolor'] = '#000000'
plt.rcParams['axes.edgecolor'] = '#FFFFFF'
plt.rcParams['axes.labelcolor'] = '#FFFFFF'
plt.rcParams['xtick.color'] = '#FFFFFF'
plt.rcParams['ytick.color'] = '#FFFFFF'
plt.rcParams['text.color'] = '#FFFFFF'
plt.rcParams['font.size'] = 14
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['lines.linewidth'] = 2.0
plt.rcParams['grid.color'] = '#444444'

# Function to compute the sum of first n integers directly
def direct_sum(n):
    return sum(range(1, n + 1))

# Function using the closed-form formula
def closed_form_sum(n):
    return n * (n + 1) // 2

# Generate data for demonstration
n_values = np.arange(1, 21)
direct_sums = [direct_sum(n) for n in n_values]
closed_form_sums = [closed_form_sum(n) for n in n_values]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(n_values, direct_sums, label="Direct Sum (Iterative)", marker="o")
plt.plot(n_values, closed_form_sums, label="Closed Form", linestyle="--", marker="x")
plt.title("Demonstration of Strong Induction: Sum of First n Integers")
plt.xlabel("n")
plt.ylabel("Sum of First n Integers")
plt.legend()
plt.grid()
plt.show()
```
