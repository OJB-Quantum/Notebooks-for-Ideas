```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.dpi'] = 500

# Create a graph
G = nx.Graph()

# Add nodes
G.add_node("Permanent Magnets")
G.add_node("Microspeakers")
G.add_node("Magnetic Fields")
# ... add other nodes as needed

# Add edges with labels
G.add_edge("Permanent Magnets", "Magnetic Fields", label='generate')
G.add_edge("Magnetic Fields", "Microspeakers", label='induce sound')
# ... add other edges as needed

# Draw the network
pos = nx.spring_layout(G)  # positions for all nodes
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=4000, font_size=10, font_weight='bold')
edge_labels = nx.get_edge_attributes(G,'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Show the plot
plt.show()
```
