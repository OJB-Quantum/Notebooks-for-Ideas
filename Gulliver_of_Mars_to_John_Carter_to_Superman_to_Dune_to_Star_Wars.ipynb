```python
"""Generate a comprehensive Graphviz lineage diagram mapping planetary romance fiction.
"""

from typing import Final

import graphviz
import IPython.display
import matplotlib as mpl


# Control knobs for diagram content, aesthetics, and rendering adjustments.
DIAGRAM_TITLE: Final[str] = "The Lineage of Planetary Romance"
DIAGRAM_DPI: Final[str] = "250"
OUTPUT_FORMAT: Final[str] = "png"

LAYOUT_DIRECTION: Final[str] = "TB"
GRAPH_BACKGROUND_COLOR: Final[str] = "white"
GRAPH_MARGIN: Final[str] = "0.20"
NODE_SEPARATION: Final[str] = "0.45"
RANK_SEPARATION: Final[str] = "0.85"

NODE_FONT: Final[str] = "Tahoma"
NODE_SHAPE: Final[str] = "box"
NODE_STYLE: Final[str] = "rounded,filled"
NODE_FILL_COLOR: Final[str] = "#f2f2f2"
NODE_BORDER_COLOR: Final[str] = "#404040"
NODE_FONT_COLOR: Final[str] = "#111111"

EDGE_COLOR: Final[str] = "#404040"
EDGE_FONT: Final[str] = "Tahoma"
EDGE_FONT_SIZE: Final[str] = "10"

# Establish the default plot resolution globally utilizing runtime configurations
mpl.rcParams['figure.dpi'] = 250
mpl.rcParams['font.sans-serif'] = ['Tahoma', 'DejaVu Sans']
mpl.rcParams['font.weight'] = 'normal'


_NODE_LABELS: Final[tuple[tuple[str, str], ...]] = (
    (
        "gulliver_of_mars",
        "Gulliver of Mars\n"
        "Edwin Lester Arnold, 1905\n"
        "Originally Lieut. Gulliver Jones: His Vacation",
    ),
    (
        "princess_of_mars",
        "A Princess of Mars/ John Carter\n"
        "Edgar Rice Burroughs, 1912\n"
        "Barsoom planetary romance template\n"
        "Leaping in a new gravity",
    ),
    (
        "superman",
        "Superman\n"
        "Jerry Siegel and Joe Shuster, 1938\n"
        "Alien strength and leaping in a new gravity regime",
    ),
    (
        "dune",
        "Dune\n"
        "Frank Herbert, 1965\n"
        "Desert planet, ecology, empire, and culture",
    ),
    (
        "star_wars",
        "Star Wars\n"
        "George Lucas, 1977\n"
        "Pulp planetary adventure and cinematic mythmaking",
    ),
)

_LINEAGE_EDGES: Final[tuple[tuple[str, str], ...]] = (
    (
        "gulliver_of_mars",
        "princess_of_mars",
    ),
    (
        "princess_of_mars",
        "superman",
    ),
    (
        "princess_of_mars",
        "dune",
    ),
    (
        "princess_of_mars",
        "star_wars",
    ),
)


def create_lineage_graph() -> graphviz.Digraph:
    """Build the planetary romance lineage graph framework.

    Returns:
        A configured Graphviz directed graph containing the lineage nodes and
        contextual influence edges.
    """
    lineage_graph = graphviz.Digraph(
        name="planetary_romance_lineage",
        comment="Planetary Romance Lineage",
        format=OUTPUT_FORMAT,
    )

    lineage_graph.attr(
        bgcolor=GRAPH_BACKGROUND_COLOR,
        dpi=DIAGRAM_DPI,
        labelloc="t",
        label=DIAGRAM_TITLE,
        margin=GRAPH_MARGIN,
        nodesep=NODE_SEPARATION,
        rankdir=LAYOUT_DIRECTION,
        ranksep=RANK_SEPARATION,
        splines="ortho",
    )
    lineage_graph.attr(
        "node",
        color=NODE_BORDER_COLOR,
        fillcolor=NODE_FILL_COLOR,
        fontcolor=NODE_FONT_COLOR,
        fontname=NODE_FONT,
        shape=NODE_SHAPE,
        style=NODE_STYLE,
    )
    lineage_graph.attr(
        "edge",
        color=EDGE_COLOR,
        fontname=EDGE_FONT,
        fontsize=EDGE_FONT_SIZE,
    )

    for node_id, label in _NODE_LABELS:
        lineage_graph.node(node_id, label)

    with lineage_graph.subgraph(name="descendant_rank") as descendant_rank:
        descendant_rank.attr(rank="same")
        descendant_rank.node("superman")
        descendant_rank.node("dune")
        descendant_rank.node("star_wars")

    for source_id, target_id in _LINEAGE_EDGES:
        lineage_graph.edge(source_id, target_id)

    return lineage_graph


def display_lineage_graph(lineage_graph: graphviz.Digraph) -> None:
    """Render and display the Graphviz graph inline dynamically.

    Args:
        lineage_graph: The Graphviz directed graph required for visual rendering.
    """
    png_data = lineage_graph.pipe(format=OUTPUT_FORMAT)
    IPython.display.display(IPython.display.Image(data=png_data))


def main() -> None:
    """Create and display the planetary romance lineage visual graphic."""
    lineage_graph = create_lineage_graph()
    display_lineage_graph(lineage_graph)


if __name__ == "__main__":
    main()
```
