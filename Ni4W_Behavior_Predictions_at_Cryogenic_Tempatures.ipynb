### Primer on Cryogenic Spin Orbit Torque Extrapolation

The provided script establishes a sophisticated mathematical framework for extrapolating the operational characteristics of Nickel Tungsten (Ni4W) Spin Orbit Torque heterostructures from standard room temperature measurements down to deep cryogenic environments. Because emergent computing paradigms such as quantum processors operate at millikelvin temperatures, researchers must comprehensively understand how magnetic memory components behave when subjected to extreme thermal deprivation. The code effectively constructs a phenomenological bridge linking established macroscopic observations at 300 K to the microphysical transport phenomena governing electron and spin dynamics at temperatures approaching absolute zero.

Understanding spin orbit torque efficiency and device switching characteristics at ultra low temperatures requires an accurate physical framework. The following material establishes a semi empirical methodology mapping room temperature variables to the cryogenic regime.


A cryogenic extrapolation remains fundamentally feasible, although researchers treat it as a semi empirical model since low temperature transport, Spin Torque Ferromagnetic Resonance (ST FMR), Second Harmonic Hall (SHH), and switching data require measurement on identical stacks. The primary literature supplies robust room temperature anchors because the $\text{Ni}_4\text{W}(211)$ crystal orientation generates conventional Y spin, out of plane Z spin, and Dresselhaus like X spin components. The room temperature $\text{Ni}_4\text{W}$ resistivity measures 150 μΩ cm, whereas the 5 nm $\text{Ni}_4\text{W}/\text{CoFeB}$ stack yields $\theta^Y_{DL}=0.727$, $\theta^X_{DL}=0.091$, and $\theta^Z_{DL}=-0.022$. The 30 nm bulk like $\text{Ni}_4\text{W}$ stack retains $\theta^Y_{DL}\approx0.3$, while the 10 nm $\text{Ni}_4\text{W}/\text{Co}$ device exhibits field free switching near $1.7 \times 10^7 \text{ A cm}^{-2}$ at room temperature. The published experimental data also demonstrate that the Density Functional Theory (DFT) spin Hall angle varies strongly with the Fermi energy level, especially near $E_F$ minus 0.33 eV, which indicates composition and doping exert greater influence than ordinary cryogenic Fermi broadening below 10 K.

The central simplification asserts that the intrinsic Berry curvature contribution to the spin Hall conductivity becomes nearly temperature independent once the thermal energy $k_BT$ falls well below the energy scale over which the DFT spin Hall conductivity fluctuates. At 10 K, $k_BT$ approximates 0.86 meV, and at 10 mK it reaches only 0.86 μeV, whereas the theoretical DFT trends span energy scales of tenths of an eV. The dominant cryogenic variations therefore originate from resistivity, spin diffusion length, interface transparency, magnetic damping, anisotropy, saturation magnetization, thermal activation, and device self heating.

A highly practical working model defines the temperature dependent resistivity $\rho(T)$ as

$$
\rho(T)=\rho_0+\left[\rho_{300}-\rho_0-A_{ee}(300)^2\right]\frac{B_5(T,\Theta_D)}{B_5(300,\Theta_D)}+A_{ee}T^2
$$

where $\rho_0$ represents the residual resistivity, $A_{ee}$ serves as the electron electron scattering coefficient, $\Theta_D$ indicates the Debye temperature, and the Bloch Grüneisen integral $B_5(T,\Theta_D)$ takes the exact mathematical form

$$
B_5(T,\Theta_D)=\left(\frac{T}{\Theta_D}\right)^5\int_0^{\Theta_D/T}\frac{x^5}{(e^x-1)(1-e^{-x})}dx
$$

The framework models the spin torque efficiency $\theta_p(T,t)$ for a specific spatial spin component $p \in \{X,Y,Z\}$ and layer thickness $t$ using the analytical expression

$$
\theta_p(T,t)=\theta_{p,\text{bulk}}(300,t)\frac{\rho(T)}{\rho_{300}}\frac{1-\operatorname{sech}[t/\lambda_s(T)]}{1-\operatorname{sech}[t/\lambda_s(300)]}\frac{\sigma^p_{\text{SH}}(T)}{\sigma^p_{\text{SH}}(300)}+\theta_{p,\text{int}}(300)\left[\frac{\rho(T)}{\rho_{300}}\right]^\eta
$$

with the temperature dependent spin diffusion length $\lambda_s(T)$ obeying the Elliott Yafet scaling relation characterized by exponent $\nu$

$$
\lambda_s(T)=\lambda_s(300)\left[\frac{\rho_{300}}{\rho(T)}\right]^\nu
$$

For the theoretical switching estimate, scientists apply a calibrated macrospin scale where the switching current density $J_{\text{sw}}(T,\tau)$ incorporates the pulse duration $\tau$

$$
J_{\text{sw}}(T,\tau)=\frac{J_{\text{sw}}(300)}{F_{\text{th}}(300,\tau)}\frac{M_s(T)H_k(T)\alpha(T)}{M_s(300)H_k(300)\alpha(300)}\frac{|\theta_{\text{eff}}(300)|}{|\theta_{\text{eff}}(T)|}F_{\text{th}}(T,\tau)
$$

where the thermal activation factor $F_{\text{th}}(T,\tau)$ relies on the energy barrier $E_B(T)$ and the attempt time $\tau_0$

$$
F_{\text{th}}(T,\tau)=\max\left[0, 1-\frac{k_BT}{E_B(T)}\ln\left(\frac{\tau}{\tau_0}\right)\right]
$$

Assuming a conservative configuration consisting of $\rho_0=0.85\rho_{300}$, $\lambda_s(300)=2 \text{ nm}$, Elliott Yafet scaling with $\nu=1$, a constant 5 nm interfacial contribution, and a 60 $k_BT$ room temperature barrier, the predicted cryogenic values collapse rapidly below 10 K because phonon scattering and thermal activation completely freeze out.

The underlying W seed layer demands special attention at 10 mK. The original $\text{Ni}_4\text{W}$ publication implements a W seed layer where X ray Diffraction (XRD) positively identifies a W(110) structural feature within the stack. Tungsten thin films frequently exhibit superconducting transitions that depend intensely on phase, residual stress, local disorder, and underlayer composition. The National Institute of Standards and Technology (NIST) reports $\alpha \text{ W}$ transitions near 15 mK and $\beta \text{ W}$ transitions between 1 and 4 K, while recent Physical Review B findings demonstrate that superconductivity in W thin films occasionally arises from amorphous W related phases near 2 to 5 K. This phenomenon implies the 10 mK and potentially 50 mK device level behavior fundamentally requires a W seed shunting or superconducting proximity check, even if the intrinsic $\text{Ni}_4\text{W}$ mathematical model remains entirely normal metallic.

```python
"""Cryogenic resistivity submodel for Ni4W transport extrapolation.

This cell estimates the temperature dependent Ni4W resistivity using a
Bloch Grueneisen phonon term, a residual resistivity term, and an optional
electron electron scattering term. The output is intended to feed later
spin orbit torque and magnonic heterostructure calculations.
"""

from __future__ import annotations

from typing import Any

from IPython.display import display
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


# ---------------------------------------------------------------------------
# GLOBAL CONTROL KNOBS
# ---------------------------------------------------------------------------
MPL_DPI = 200

TEMPERATURES_K = np.array([300.0, 10.0, 4.0, 1.0, 0.050, 0.010])
PLOT_TEMPERATURES_K = np.geomspace(0.010, 300.0, 600)

RHO_300_UOHM_CM = 150.0
RESIDUAL_RHO_FRACTION = 0.85
DEBYE_TEMPERATURE_K = 350.0
ELECTRON_ELECTRON_A_UOHM_CM_K2 = 0.0

BG_GRID_POINTS = 40_000
BG_X_MAX = 80.0

MAKE_SENSITIVITY_PLOT = True
RESIDUAL_RHO_FRACTIONS_FOR_SWEEP = np.array([0.50, 0.70, 0.85, 0.95])
# ---------------------------------------------------------------------------


mpl.rcParams["figure.dpi"] = MPL_DPI
mpl.rcParams["savefig.dpi"] = MPL_DPI

try:
    import cupy as cp

    if cp.cuda.runtime.getDeviceCount() > 0:
        XP: Any = cp
        USING_CUPY = True
    else:
        XP = np
        USING_CUPY = False
except Exception:
    XP = np
    USING_CUPY = False


def to_numpy(values: Any) -> np.ndarray:
    """Return values as a NumPy array for Pandas and Matplotlib."""

    if USING_CUPY:
        return XP.asnumpy(values)
    return np.asarray(values)


def trapezoid(values: Any, grid: Any, axis: int) -> Any:
    """Evaluate a trapezoidal integral with NumPy or CuPy."""

    if hasattr(XP, "trapezoid"):
        return XP.trapezoid(values, grid, axis=axis)
    return XP.trapz(values, grid, axis=axis)


def bloch_grueneisen_factor(temperatures_k: Any) -> Any:
    """Compute the dimensionless Bloch Grueneisen transport factor."""

    temperatures = XP.atleast_1d(XP.asarray(temperatures_k, dtype=XP.float64))
    x_grid = XP.linspace(1.0e-9, BG_X_MAX, BG_GRID_POINTS)

    upper_limit = DEBYE_TEMPERATURE_K / temperatures[:, None]
    active_mask = x_grid[None, :] <= upper_limit

    sinh_term = XP.sinh(0.5 * x_grid)
    integrand = 0.25 * x_grid**5 / sinh_term**2
    integral = trapezoid(integrand[None, :] * active_mask, x_grid, axis=1)

    return (temperatures / DEBYE_TEMPERATURE_K) ** 5 * integral


def resistivity_uohm_cm(
    temperatures_k: Any,
    residual_rho_fraction: float = RESIDUAL_RHO_FRACTION,
) -> Any:
    """Evaluate the Ni4W resistivity in micro ohm cm."""

    temperatures = XP.atleast_1d(XP.asarray(temperatures_k, dtype=XP.float64))

    rho_0 = residual_rho_fraction * RHO_300_UOHM_CM
    rho_ee_300 = ELECTRON_ELECTRON_A_UOHM_CM_K2 * 300.0**2
    rho_ph_300 = RHO_300_UOHM_CM - rho_0 - rho_ee_300
    rho_ph_300 = max(float(rho_ph_300), 0.0)

    bg_values = bloch_grueneisen_factor(temperatures)
    bg_300 = bloch_grueneisen_factor(XP.asarray([300.0], dtype=XP.float64))[0]
    rho_ee = ELECTRON_ELECTRON_A_UOHM_CM_K2 * temperatures**2

    return rho_0 + rho_ph_300 * bg_values / bg_300 + rho_ee


def build_resistivity_table() -> pd.DataFrame:
    """Build a table for the target cryogenic temperatures."""

    temperatures = XP.asarray(TEMPERATURES_K, dtype=XP.float64)
    rho = resistivity_uohm_cm(temperatures)
    rho_ratio = rho / RHO_300_UOHM_CM

    return pd.DataFrame(
        {
            "T_K": to_numpy(temperatures),
            "rho_uohm_cm": to_numpy(rho),
            "rho_over_rho_300": to_numpy(rho_ratio),
        }
    )


def plot_resistivity_curve() -> None:
    """Plot the baseline Ni4W resistivity curve."""

    temperatures = XP.asarray(PLOT_TEMPERATURES_K, dtype=XP.float64)
    rho = resistivity_uohm_cm(temperatures)

    plt.figure(figsize=(8, 5))
    plt.plot(to_numpy(temperatures), to_numpy(rho), linewidth=2.0)
    plt.xscale("log")
    plt.xlabel("Temperature (K)")
    plt.ylabel(r"Resistivity $\rho$ ($\mu\Omega$ cm)")
    plt.title(r"Cryogenic extrapolation of Ni$_4$W resistivity")
    plt.grid(True, which="both", alpha=0.3)
    plt.tight_layout()
    plt.show()


def plot_residual_fraction_sweep() -> None:
    """Plot sensitivity to the assumed residual resistivity fraction."""

    temperatures = XP.asarray(PLOT_TEMPERATURES_K, dtype=XP.float64)

    plt.figure(figsize=(8, 5))
    for residual_fraction in RESIDUAL_RHO_FRACTIONS_FOR_SWEEP:
        rho = resistivity_uohm_cm(
            temperatures,
            residual_rho_fraction=float(residual_fraction),
        )
        label = rf"$\rho_0/\rho_{{300}}={residual_fraction:.2f}$"
        plt.plot(to_numpy(temperatures), to_numpy(rho), label=label)

    plt.xscale("log")
    plt.xlabel("Temperature (K)")
    plt.ylabel(r"Resistivity $\rho$ ($\mu\Omega$ cm)")
    plt.title(r"Residual resistivity sensitivity for Ni$_4$W")
    plt.grid(True, which="both", alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()


def main() -> pd.DataFrame:
    """Run the resistivity submodel and render notebook outputs."""

    backend = "CuPy GPU" if USING_CUPY else "NumPy CPU"
    print(f"Backend: {backend}")

    table = build_resistivity_table()
    display(table)

    plot_resistivity_curve()

    if MAKE_SENSITIVITY_PLOT:
        plot_residual_fraction_sweep()

    return table


if __name__ == "__main__":
    resistivity_results = main()
```

```python
"""Cryogenic extrapolation model for Ni4W spin orbit torque stacks.

The model is anchored to room temperature quantities reported for Ni4W
heterostructures and extrapolates resistivity, spin diffusion, torque
efficiency, and a calibrated switching current proxy.
"""

from __future__ import annotations

from dataclasses import dataclass
import math
from typing import Any

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ---------------------------------------------------------------------------
# GLOBAL CONTROL KNOBS (Parameter Definitions)
# ---------------------------------------------------------------------------
TEMPERATURES_K = np.array([300.0, 10.0, 4.0, 1.0, 0.050, 0.010])

RHO_NI4W_300_UOHM_CM = 150.0
RESIDUAL_RHO_FRACTION = 0.85
DEBYE_TEMPERATURE_K = 350.0
ELECTRON_ELECTRON_A_UOHM_CM_K2 = 0.0

SPIN_DIFFUSION_LENGTH_300_NM = 2.0
SPIN_DIFFUSION_EXPONENT = 1.0
SPIN_DIFFUSION_LENGTH_MAX_NM = 20.0

INTERFACE_RHO_EXPONENT = 0.0
SOMMERFELD_CURVATURE_EV_NEG2 = 0.0

BG_GRID_POINTS = 40_000
BG_X_MAX = 80.0

SWITCHING_J_300_A_CM2 = 1.7e7
SWITCHING_THETA_Y_300 = 0.52
SWITCHING_STACK_THICKNESS_NM = 10.0
PULSE_DURATION_S = 10e-3
ATTEMPT_TIME_S = 1e-9
BARRIER_KBT_AT_300 = 60.0
LOW_TEMP_MS_RATIO = 1.05
LOW_TEMP_HK_RATIO = 1.10
LOW_TEMP_ALPHA_RATIO = 0.80

MPL_DPI = 200
# ---------------------------------------------------------------------------

@dataclass(frozen=True)
class TorqueCase:
    """Room temperature anchor for one torque efficiency component."""

    name: str
    thickness_nm: float
    theta_bulk_300: float
    theta_interface_300: float = 0.0
    curvature_ev_neg2: float = SOMMERFELD_CURVATURE_EV_NEG2


TORQUE_CASES = (
    TorqueCase(
        name="5 nm Ni4W, Y damping like total",
        thickness_nm=5.0,
        theta_bulk_300=0.300,
        theta_interface_300=0.727 - 0.300,
    ),
    TorqueCase(
        name="5 nm Ni4W, X damping like",
        thickness_nm=5.0,
        theta_bulk_300=0.091,
        theta_interface_300=0.0,
    ),
    TorqueCase(
        name="5 nm Ni4W, Z damping like",
        thickness_nm=5.0,
        theta_bulk_300=-0.022,
        theta_interface_300=0.0,
    ),
    TorqueCase(
        name="30 nm Ni4W, Y damping like bulk",
        thickness_nm=30.0,
        theta_bulk_300=0.300,
        theta_interface_300=0.0,
    ),
    TorqueCase(
        name="10 nm Ni4W, Y switching proxy",
        thickness_nm=SWITCHING_STACK_THICKNESS_NM,
        theta_bulk_300=SWITCHING_THETA_Y_300,
        theta_interface_300=0.0,
    ),
)

mpl.rcParams["figure.dpi"] = MPL_DPI
mpl.rcParams["savefig.dpi"] = MPL_DPI

try:
    import cupy as cp

    if cp.cuda.runtime.getDeviceCount() > 0:
        XP: Any = cp
        USING_CUPY = True
    else:
        XP = np
        USING_CUPY = False
except Exception:
    XP = np
    USING_CUPY = False


K_B_EV_PER_K = 8.617333262145e-5


def to_numpy(values: Any) -> np.ndarray:
    """Return a NumPy array from either a CuPy or NumPy array."""

    if USING_CUPY:
        return XP.asnumpy(values)
    return np.asarray(values)


def trapezoid(values: Any, grid: Any, axis: int) -> Any:
    """Evaluate a trapezoidal integral with NumPy or CuPy."""

    if hasattr(XP, "trapezoid"):
        return XP.trapezoid(values, grid, axis=axis)
    return XP.trapz(values, grid, axis=axis)


def bloch_gruneisen_factor(temperatures_k: Any) -> Any:
    """Compute the dimensionless Bloch Grueneisen transport factor."""

    temperatures = XP.asarray(temperatures_k, dtype=XP.float64)
    x_grid = XP.linspace(1.0e-9, BG_X_MAX, BG_GRID_POINTS)
    x_max = DEBYE_TEMPERATURE_K / temperatures[:, None]
    active = x_grid[None, :] <= x_max

    sinh_term = XP.sinh(0.5 * x_grid)
    integrand = 0.25 * x_grid**5 / sinh_term**2
    integral = trapezoid(integrand[None, :] * active, x_grid, axis=1)

    return (temperatures / DEBYE_TEMPERATURE_K) ** 5 * integral


def resistivity_uohm_cm(temperatures_k: Any) -> Any:
    """Evaluate Ni4W resistivity in micro ohm cm."""

    temperatures = XP.asarray(temperatures_k, dtype=XP.float64)
    rho_0 = RESIDUAL_RHO_FRACTION * RHO_NI4W_300_UOHM_CM
    rho_ee_300 = ELECTRON_ELECTRON_A_UOHM_CM_K2 * 300.0**2
    rho_ph_300 = RHO_NI4W_300_UOHM_CM - rho_0 - rho_ee_300
    rho_ph_300 = max(float(rho_ph_300), 0.0)

    bg_values = bloch_gruneisen_factor(temperatures)
    bg_300 = bloch_gruneisen_factor(
        XP.asarray([300.0], dtype=XP.float64)
    )[0]
    rho_ee = ELECTRON_ELECTRON_A_UOHM_CM_K2 * temperatures**2

    return rho_0 + rho_ph_300 * bg_values / bg_300 + rho_ee


def spin_diffusion_length_nm(rho_uohm_cm: Any) -> Any:
    """Estimate spin diffusion length using Elliott Yafet scaling."""

    rho_ratio = RHO_NI4W_300_UOHM_CM / rho_uohm_cm
    lambda_s = SPIN_DIFFUSION_LENGTH_300_NM * (rho_ratio**SPIN_DIFFUSION_EXPONENT)
    return XP.minimum(lambda_s, SPIN_DIFFUSION_LENGTH_MAX_NM)


def spin_absorption(thickness_nm: float, lambda_s_nm: Any) -> Any:
    """Return the finite thickness spin current absorption factor."""

    return 1.0 - 1.0 / XP.cosh(thickness_nm / lambda_s_nm)


def sommerfeld_ratio(temperatures_k: Any, curvature_ev_neg2: float) -> Any:
    """Estimate intrinsic spin Hall conductivity thermal smearing."""

    temperatures = XP.asarray(temperatures_k, dtype=XP.float64)
    coefficient = (math.pi**2 / 6.0) * K_B_EV_PER_K**2
    delta_t2 = temperatures**2 - 300.0**2
    return 1.0 + coefficient * curvature_ev_neg2 * delta_t2


def theta_temperature(case: TorqueCase, temperatures_k: Any) -> Any:
    """Evaluate a signed SOT efficiency for one anchored torque case."""

    temperatures = XP.asarray(temperatures_k, dtype=XP.float64)
    rho = resistivity_uohm_cm(temperatures)
    rho_ratio = rho / RHO_NI4W_300_UOHM_CM

    lambda_s = spin_diffusion_length_nm(rho)
    absorption = spin_absorption(case.thickness_nm, lambda_s)
    absorption_300 = spin_absorption(
        case.thickness_nm,
        XP.asarray([SPIN_DIFFUSION_LENGTH_300_NM], dtype=XP.float64),
    )[0]

    sigma_ratio = sommerfeld_ratio(temperatures, case.curvature_ev_neg2)
    theta_bulk = case.theta_bulk_300 * rho_ratio * absorption * (sigma_ratio / absorption_300)
    theta_interface = case.theta_interface_300 * (rho_ratio**INTERFACE_RHO_EXPONENT)

    return theta_bulk + theta_interface


def saturation_ratio(
    temperatures_k: Any,
    low_temperature_ratio: float,
    exponent: float,
) -> Any:
    """Return a smooth 300 K to low temperature property ratio."""

    temperatures = XP.asarray(temperatures_k, dtype=XP.float64)
    reduced_temperature = XP.clip(temperatures / 300.0, 0.0, 1.0)
    return low_temperature_ratio - (
        low_temperature_ratio - 1.0
    ) * reduced_temperature**exponent


def thermal_activation_factor(temperatures_k: Any, barrier_ratio: Any) -> Any:
    """Compute the thermal switching assist factor for a pulse."""

    temperatures = XP.asarray(temperatures_k, dtype=XP.float64)
    log_factor = math.log(PULSE_DURATION_S / ATTEMPT_TIME_S)
    factor = 1.0 - (
        temperatures / 300.0
    ) * log_factor / (BARRIER_KBT_AT_300 * barrier_ratio)
    return XP.clip(factor, 0.0, 1.0)


def switching_current_proxy_a_cm2(temperatures_k: Any) -> tuple[Any, Any]:
    """Estimate a calibrated switching current and ohmic power ratio."""

    temperatures = XP.asarray(temperatures_k, dtype=XP.float64)
    switching_case = TorqueCase(
        name="switching",
        thickness_nm=SWITCHING_STACK_THICKNESS_NM,
        theta_bulk_300=SWITCHING_THETA_Y_300,
    )

    theta = theta_temperature(switching_case, temperatures)
    theta_ratio = XP.abs(theta / SWITCHING_THETA_Y_300)

    ms_ratio = saturation_ratio(temperatures, LOW_TEMP_MS_RATIO, 1.5)
    hk_ratio = saturation_ratio(temperatures, LOW_TEMP_HK_RATIO, 2.0)
    alpha_ratio = saturation_ratio(temperatures, LOW_TEMP_ALPHA_RATIO, 2.0)
    barrier_ratio = ms_ratio * hk_ratio

    thermal_factor = thermal_activation_factor(temperatures, barrier_ratio)
    thermal_factor_300 = thermal_activation_factor(
        XP.asarray([300.0], dtype=XP.float64),
        XP.asarray([1.0], dtype=XP.float64),
    )[0]

    intrinsic_j_300 = SWITCHING_J_300_A_CM2 / thermal_factor_300
    switching_j = intrinsic_j_300 * ms_ratio * hk_ratio * alpha_ratio * (thermal_factor / theta_ratio)

    rho = resistivity_uohm_cm(temperatures)
    rho_ratio = rho / RHO_NI4W_300_UOHM_CM
    power_ratio = rho_ratio * (switching_j / SWITCHING_J_300_A_CM2) ** 2

    return switching_j, power_ratio


def build_results_table() -> pd.DataFrame:
    """Build a tidy table for the target cryogenic temperatures."""

    temperatures = XP.asarray(TEMPERATURES_K, dtype=XP.float64)
    rho = resistivity_uohm_cm(temperatures)
    lambda_s = spin_diffusion_length_nm(rho)
    switching_j, power_ratio = switching_current_proxy_a_cm2(temperatures)

    rows = []
    for index, temperature in enumerate(to_numpy(temperatures)):
        row = {
            "T_K": temperature,
            "rho_uohm_cm": to_numpy(rho)[index],
            "lambda_s_nm": to_numpy(lambda_s)[index],
            "J_sw_proxy_A_cm2": to_numpy(switching_j)[index],
            "ohmic_power_ratio": to_numpy(power_ratio)[index],
        }
        for case in TORQUE_CASES:
            theta_values = theta_temperature(case, temperatures)
            row[case.name] = to_numpy(theta_values)[index]
        rows.append(row)

    return pd.DataFrame(rows)


def plot_results(table: pd.DataFrame) -> None:
    """Render plots in the active Colab notebook without saving files."""

    plt.figure()
    plt.loglog(table["T_K"], table["rho_uohm_cm"], marker="o")
    plt.gca().invert_xaxis()
    plt.xlabel("Temperature (K)")
    plt.ylabel(r"Ni$_4$W resistivity ($\mu\Omega$ cm)")
    plt.title("Resistivity extrapolation")
    plt.grid(True, which="both", alpha=0.3)
    plt.show()

    theta_columns = [case.name for case in TORQUE_CASES]
    plt.figure()
    for column in theta_columns:
        plt.semilogx(table["T_K"], table[column], marker="o", label=column)
    plt.gca().invert_xaxis()
    plt.xlabel("Temperature (K)")
    plt.ylabel("SOT efficiency")
    plt.title("Torque efficiency extrapolation")
    plt.grid(True, which="both", alpha=0.3)
    plt.legend(fontsize=7)
    plt.show()

    plt.figure()
    plt.semilogx(table["T_K"], table["J_sw_proxy_A_cm2"], marker="o")
    plt.gca().invert_xaxis()
    plt.xlabel("Temperature (K)")
    plt.ylabel(r"Switching current density (A cm$^{-2}$)")
    plt.title("Calibrated switching current proxy")
    plt.grid(True, which="both", alpha=0.3)
    plt.show()


def main() -> pd.DataFrame:
    """Run the cryogenic extrapolation and show the output table."""

    backend = "CuPy GPU" if USING_CUPY else "NumPy CPU"
    print(f"Backend: {backend}")

    table = build_results_table()
    display_columns = [
        "T_K",
        "rho_uohm_cm",
        "lambda_s_nm",
        "5 nm Ni4W, Y damping like total",
        "5 nm Ni4W, Z damping like",
        "30 nm Ni4W, Y damping like bulk",
        "J_sw_proxy_A_cm2",
        "ohmic_power_ratio",
    ]
    print(table[display_columns].to_string(index=False))
    plot_results(table)
    return table


if __name__ == "__main__":
    results = main()
```

```python
"""
High throughput cryogenic extrapolation model for Spin Orbit Torque materials.

This implementation vectorizes the physical transport equations across a massive
two dimensional parameter space while isolating the integration dimension to
prevent video memory overflow on the computing accelerator.
"""

from __future__ import annotations

from dataclasses import dataclass
import math
from typing import Any

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ---------------------------------------------------------------------------
# GLOBAL CONTROL KNOBS (Parameter Definitions)
# ---------------------------------------------------------------------------
# Generate a massive parameter space safely within 24GB VRAM limits
TEMPERATURE_POINTS = 2000
RESISTIVITY_POINTS = 2000

TEMP_MIN_K = 0.010
TEMP_MAX_K = 300.0
RESIDUAL_FRACTION_MIN = 0.40
RESIDUAL_FRACTION_MAX = 0.95

RHO_NI4W_300_UOHM_CM = 150.0
DEBYE_TEMPERATURE_K = 350.0
ELECTRON_ELECTRON_A_UOHM_CM_K2 = 0.0

SPIN_DIFFUSION_LENGTH_300_NM = 2.0
SPIN_DIFFUSION_EXPONENT = 1.0
SPIN_DIFFUSION_LENGTH_MAX_NM = 20.0

INTERFACE_RHO_EXPONENT = 0.0
SOMMERFELD_CURVATURE_EV_NEG2 = 0.0

BG_GRID_POINTS = 5_000
BG_X_MAX = 80.0

SWITCHING_J_300_A_CM2 = 1.7e7
SWITCHING_THETA_Y_300 = 0.52
SWITCHING_STACK_THICKNESS_NM = 10.0
PULSE_DURATION_S = 10e-3
ATTEMPT_TIME_S = 1e-9
BARRIER_KBT_AT_300 = 60.0

MPL_RESOLUTION_DPI = 250
# ---------------------------------------------------------------------------

@dataclass(frozen=True)
class TorqueCase:
    """Room temperature anchor defining one specific torque component."""

    name: str
    thickness_nm: float
    theta_bulk_300: float
    theta_interface_300: float = 0.0
    curvature_ev_neg2: float = SOMMERFELD_CURVATURE_EV_NEG2


SWITCHING_CASE = TorqueCase(
    name="10 nm Ni4W, Y switching proxy",
    thickness_nm=SWITCHING_STACK_THICKNESS_NM,
    theta_bulk_300=SWITCHING_THETA_Y_300,
    theta_interface_300=0.0,
)

# Enforce strict Nature style publication graphics requirements
mpl.rcParams["figure.dpi"] = MPL_RESOLUTION_DPI
mpl.rcParams["savefig.dpi"] = MPL_RESOLUTION_DPI
mpl.rcParams["font.family"] = "serif"

try:
    import cupy as cp

    if cp.cuda.runtime.getDeviceCount() > 0:
        XP: Any = cp
        USING_CUPY = True
    else:
        XP = np
        USING_CUPY = False
except Exception:
    XP = np
    USING_CUPY = False


K_B_EV_PER_K = 8.617333262145e-5


def to_numpy(values: Any) -> np.ndarray:
    """Return a standard NumPy array by detaching memory from the device."""
    if USING_CUPY:
        return XP.asnumpy(values)
    return np.asarray(values)


def trapezoid(values: Any, grid: Any, axis: int) -> Any:
    """Evaluate a numerical integral utilizing the designated processor."""
    if hasattr(XP, "trapezoid"):
        return XP.trapezoid(values, grid, axis=axis)
    return XP.trapz(values, grid, axis=axis)


def bloch_gruneisen_factor(temperatures_k: Any) -> Any:
    """Compute the dimensionless electron phonon transport factor."""
    temperatures = XP.asarray(temperatures_k, dtype=XP.float64)
    x_grid = XP.linspace(1.0e-9, BG_X_MAX, BG_GRID_POINTS)

    x_max = DEBYE_TEMPERATURE_K / temperatures[..., None]
    active = x_grid > 0.0
    active_mask = active & (x_grid <= x_max)

    sinh_term = XP.sinh(0.5 * x_grid)
    integrand = 0.25 * x_grid**5 / sinh_term**2
    integral = trapezoid(integrand * active_mask, x_grid, axis=-1)

    return (temperatures / DEBYE_TEMPERATURE_K) ** 5 * integral


def resistivity_uohm_cm(temp_array: Any, frac_array: Any) -> tuple[Any, Any, Any]:
    """Evaluate the complete intrinsic resistivity tensor over the parameter grid."""
    temperatures = XP.asarray(temp_array, dtype=XP.float64)
    fractions = XP.asarray(frac_array, dtype=XP.float64)

    # Generate the orthogonal two-dimensional matrices for spatial plotting
    temp_grid, frac_grid = XP.meshgrid(temperatures, fractions, indexing='ij')

    rho_0 = frac_grid * RHO_NI4W_300_UOHM_CM
    rho_ee_300 = ELECTRON_ELECTRON_A_UOHM_CM_K2 * 300.0**2
    rho_ph_300 = XP.maximum(RHO_NI4W_300_UOHM_CM - rho_0 - rho_ee_300, 0.0)

    # Restrict the computationally expensive integration strictly to the one-dimensional temperature vector
    bg_values_1d = bloch_gruneisen_factor(temperatures)
    bg_300 = bloch_gruneisen_factor(XP.asarray([300.0], dtype=XP.float64))[0]

    # Broadcast the temperature-dependent vector across the residual fraction dimension
    bg_ratio_2d = bg_values_1d[:, None] / bg_300
    rho_ee_2d = ELECTRON_ELECTRON_A_UOHM_CM_K2 * (temp_grid**2)

    rho_tensor = rho_0 + rho_ph_300 * bg_ratio_2d + rho_ee_2d

    return temp_grid, frac_grid, rho_tensor


def execute_high_throughput_sweep() -> tuple[Any, Any, Any]:
    """Saturate the accelerator by computing a four million element phase space."""
    temp_array = XP.linspace(TEMP_MIN_K, TEMP_MAX_K, TEMPERATURE_POINTS)
    frac_array = XP.linspace(RESIDUAL_FRACTION_MIN, RESIDUAL_FRACTION_MAX, RESISTIVITY_POINTS)

    temp_grid, frac_grid, rho_tensor = resistivity_uohm_cm(temp_array, frac_array)

    return temp_grid, frac_grid, rho_tensor


def plot_multidimensional_results(temp_grid: Any, frac_grid: Any, rho_tensor: Any) -> None:
    """Render the saturated tensor computation directly into the active notebook."""
    temp_cpu = to_numpy(temp_grid)
    frac_cpu = to_numpy(frac_grid)
    rho_cpu = to_numpy(rho_tensor)

    plt.figure(figsize=(8, 6))
    contour = plt.contourf(
        temp_cpu,
        frac_cpu,
        rho_cpu,
        levels=50,
        cmap='viridis'
    )
    cbar = plt.colorbar(contour)
    cbar.set_label(r"Resistivity ($\mu\Omega$ cm)", fontsize=11)

    plt.xlabel("Temperature (K)", fontsize=12)
    plt.ylabel("Residual Resistivity Fraction", fontsize=12)
    plt.title("High Throughput Cryogenic Transport Phase Space", fontsize=14)
    plt.grid(True, alpha=0.2, color='white')
    plt.show()


def main() -> None:
    """Initialize the architectural pipeline and execute the phase space mapping."""
    backend_identifier = "CuPy GPU" if USING_CUPY else "NumPy CPU"
    print(f"Active Computational Backend: {backend_identifier}")

    temp_grid, frac_grid, rho_tensor = execute_high_throughput_sweep()
    print(f"Computed Tensor Dimensions: {rho_tensor.shape}")
    print(f"Total Evaluated Thermodynamic States: {rho_tensor.size:,}")

    plot_multidimensional_results(temp_grid, frac_grid, rho_tensor)


if __name__ == "__main__":
    main()
```

```python
"""GPU accelerated magnonic calculations for Ni4W heterostructures.

This notebook models Ni4W as a spin orbit torque transducer and models
Co or CoFeB as the magnon host. It computes cryogenic magnon occupation,
SOT antidamping thresholds, normalized drive maps, driven response spectra,
and a compact switching proxy anchored to the reported Ni4W devices.
"""

from __future__ import annotations

from dataclasses import dataclass
import math
from typing import Any

from IPython.display import display
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


# ---------------------------------------------------------------------------
# GLOBAL CONTROL KNOBS
# ---------------------------------------------------------------------------
MPL_DPI = 250

TEMPERATURES_K = np.array([300.0, 10.0, 4.0, 1.0, 0.050, 0.010])
PLOT_TEMPERATURES_K = np.array([10.0, 4.0, 1.0, 0.050, 0.010])

FREQUENCY_RANGE_GHZ = np.linspace(0.5, 80.0, 1200)
TEMPERATURE_GRID_K = np.geomspace(0.010, 300.0, 500)
FREQUENCY_GRID_GHZ = np.linspace(0.5, 100.0, 800)
CURRENT_GRID_A_CM2 = np.linspace(0.0, 5.0e7, 800)

TARGET_MODE_FREQUENCY_GHZ = 10.0
DRIVE_CURRENT_A_CM2 = 1.0e7
DRIVE_MAP_TEMPERATURE_K = 0.050
MAGNON_RESPONSE_TEMPERATURE_K = 0.050

RHO_NI4W_300_UOHM_CM = 150.0
RESIDUAL_RHO_FRACTION = 0.85
DEBYE_TEMPERATURE_K = 350.0
BG_GRID_POINTS = 30_000
BG_X_MAX = 80.0
INTERFACE_RHO_EXPONENT = 0.0

LOW_TEMP_MS_RATIO = 1.05
LOW_TEMP_HK_RATIO = 1.10
LOW_TEMP_ALPHA_RATIO = 0.80
PROPERTY_SATURATION_EXPONENT = 1.5

COFEB_MS_A_M = 1.20e6
CO_MS_A_M = 1.40e6

COFEB_THICKNESS_NM = 5.0
CO_THICKNESS_NM = 0.9

COFEB_ALPHA_300 = 0.012
CO_ALPHA_300 = 0.040

COFEB_EFFECTIVE_FIELD_300_T = 1.00
CO_EFFECTIVE_FIELD_300_T = 0.55

NI4W_5NM_THETA_Y_BULK_300 = 0.300
NI4W_5NM_THETA_Y_INTERFACE_300 = 0.727 - 0.300
NI4W_5NM_THETA_X_300 = 0.091
NI4W_5NM_THETA_Z_300 = -0.022

NI4W_30NM_THETA_Y_BULK_300 = 0.300
NI4W_30NM_THETA_Y_INTERFACE_300 = 0.0
NI4W_30NM_THETA_X_300 = 0.006
NI4W_30NM_THETA_Z_300 = -0.015

NI4W_10NM_CO_THETA_Y_BULK_300 = 0.300
NI4W_10NM_CO_THETA_Y_INTERFACE_300 = 0.220
NI4W_10NM_CO_THETA_X_300 = 0.010
NI4W_10NM_CO_THETA_Z_300 = -0.018
NI4W_CO_REPORTED_SWITCHING_A_CM2 = 1.7e7

PT_CO_THETA_Y_300 = 0.12
PT_CO_THETA_Z_300 = 0.0
PT_CO_REPORTED_SWITCHING_A_CM2 = 2.7e7
# ---------------------------------------------------------------------------
# END CONTROL KNOBS
# ---------------------------------------------------------------------------


mpl.rcParams["figure.dpi"] = MPL_DPI
mpl.rcParams["savefig.dpi"] = MPL_DPI

try:
    import cupy as cp

    if cp.cuda.runtime.getDeviceCount() > 0:
        XP: Any = cp
        USING_CUPY = True
    else:
        XP = np
        USING_CUPY = False
except Exception:
    XP = np
    USING_CUPY = False


E_CHARGE_C = 1.602176634e-19
H_J_S = 6.62607015e-34
HBAR_J_S = 1.054571817e-34
K_B_J_K = 1.380649e-23
GAMMA_RAD_S_T = 1.76085963023e11
GAMMA_GHZ_T = GAMMA_RAD_S_T / (2.0 * math.pi) / 1.0e9


@dataclass(frozen=True)
class StackCase:
    """Parameters for one spin torque driven magnetic heterostructure."""

    name: str
    host: str
    theta_y_bulk_300: float
    theta_y_interface_300: float
    theta_x_300: float
    theta_z_300: float
    ms_300_a_m: float
    ferromagnet_thickness_nm: float
    alpha_300: float
    effective_field_300_t: float
    reported_switching_a_cm2: float | None = None


STACK_CASES = (
    StackCase(
        name="Ni4W 5 nm / CoFeB",
        host="CoFeB",
        theta_y_bulk_300=NI4W_5NM_THETA_Y_BULK_300,
        theta_y_interface_300=NI4W_5NM_THETA_Y_INTERFACE_300,
        theta_x_300=NI4W_5NM_THETA_X_300,
        theta_z_300=NI4W_5NM_THETA_Z_300,
        ms_300_a_m=COFEB_MS_A_M,
        ferromagnet_thickness_nm=COFEB_THICKNESS_NM,
        alpha_300=COFEB_ALPHA_300,
        effective_field_300_t=COFEB_EFFECTIVE_FIELD_300_T,
    ),
    StackCase(
        name="Ni4W 30 nm / CoFeB",
        host="CoFeB",
        theta_y_bulk_300=NI4W_30NM_THETA_Y_BULK_300,
        theta_y_interface_300=NI4W_30NM_THETA_Y_INTERFACE_300,
        theta_x_300=NI4W_30NM_THETA_X_300,
        theta_z_300=NI4W_30NM_THETA_Z_300,
        ms_300_a_m=COFEB_MS_A_M,
        ferromagnet_thickness_nm=COFEB_THICKNESS_NM,
        alpha_300=COFEB_ALPHA_300,
        effective_field_300_t=COFEB_EFFECTIVE_FIELD_300_T,
    ),
    StackCase(
        name="Ni4W 10 nm / Co field free device",
        host="Co",
        theta_y_bulk_300=NI4W_10NM_CO_THETA_Y_BULK_300,
        theta_y_interface_300=NI4W_10NM_CO_THETA_Y_INTERFACE_300,
        theta_x_300=NI4W_10NM_CO_THETA_X_300,
        theta_z_300=NI4W_10NM_CO_THETA_Z_300,
        ms_300_a_m=CO_MS_A_M,
        ferromagnet_thickness_nm=CO_THICKNESS_NM,
        alpha_300=CO_ALPHA_300,
        effective_field_300_t=CO_EFFECTIVE_FIELD_300_T,
        reported_switching_a_cm2=NI4W_CO_REPORTED_SWITCHING_A_CM2,
    ),
    StackCase(
        name="Pt / Co assisted field reference",
        host="Co",
        theta_y_bulk_300=PT_CO_THETA_Y_300,
        theta_y_interface_300=0.0,
        theta_x_300=0.0,
        theta_z_300=PT_CO_THETA_Z_300,
        ms_300_a_m=CO_MS_A_M,
        ferromagnet_thickness_nm=0.8,
        alpha_300=CO_ALPHA_300,
        effective_field_300_t=CO_EFFECTIVE_FIELD_300_T,
        reported_switching_a_cm2=PT_CO_REPORTED_SWITCHING_A_CM2,
    ),
)


def to_numpy(values: Any) -> np.ndarray:
    """Return an array as a NumPy object for Pandas and Matplotlib."""

    if USING_CUPY:
        return XP.asnumpy(values)
    return np.asarray(values)


def trapezoid(values: Any, grid: Any, axis: int) -> Any:
    """Evaluate a trapezoidal integral with NumPy or CuPy."""

    if hasattr(XP, "trapezoid"):
        return XP.trapezoid(values, grid, axis=axis)
    return XP.trapz(values, grid, axis=axis)


def bloch_gruneisen_factor(temperatures_k: Any) -> Any:
    """Compute the dimensionless Bloch Grueneisen transport factor."""

    temperatures = XP.atleast_1d(XP.asarray(temperatures_k, dtype=XP.float64))
    x_grid = XP.linspace(1.0e-9, BG_X_MAX, BG_GRID_POINTS)
    x_max = DEBYE_TEMPERATURE_K / temperatures[:, None]
    active = x_grid[None, :] <= x_max

    sinh_term = XP.sinh(0.5 * x_grid)
    integrand = 0.25 * x_grid**5 / sinh_term**2
    integral = trapezoid(integrand[None, :] * active, x_grid, axis=1)

    return (temperatures / DEBYE_TEMPERATURE_K) ** 5 * integral


def rho_ratio_ni4w(temperatures_k: Any) -> Any:
    """Return the Ni4W resistivity ratio rho(T) over rho(300 K)."""

    temperatures = XP.atleast_1d(XP.asarray(temperatures_k, dtype=XP.float64))
    rho_0 = RESIDUAL_RHO_FRACTION * RHO_NI4W_300_UOHM_CM
    rho_ph_300 = RHO_NI4W_300_UOHM_CM - rho_0
    bg_values = bloch_gruneisen_factor(temperatures)
    bg_300 = bloch_gruneisen_factor(XP.asarray([300.0], dtype=XP.float64))[0]
    rho = rho_0 + rho_ph_300 * bg_values / bg_300
    return rho / RHO_NI4W_300_UOHM_CM


def saturation_ratio(
    temperatures_k: Any,
    low_temperature_ratio: float,
    exponent: float,
) -> Any:
    """Return a smooth 300 K to low temperature property ratio."""

    temperatures = XP.atleast_1d(XP.asarray(temperatures_k, dtype=XP.float64))
    reduced_temperature = XP.clip(temperatures / 300.0, 0.0, 1.0)
    return low_temperature_ratio - (
        low_temperature_ratio - 1.0
    ) * reduced_temperature**exponent


def theta_y_temperature(case: StackCase, temperatures_k: Any) -> Any:
    """Evaluate the temperature dependent conventional SOT efficiency."""

    rho_ratio = rho_ratio_ni4w(temperatures_k)
    theta_bulk = case.theta_y_bulk_300 * rho_ratio
    theta_interface = case.theta_y_interface_300
    theta_interface *= rho_ratio**INTERFACE_RHO_EXPONENT
    return theta_bulk + theta_interface


def theta_z_temperature(case: StackCase, temperatures_k: Any) -> Any:
    """Evaluate the temperature dependent out of plane SOT efficiency."""

    rho_ratio = rho_ratio_ni4w(temperatures_k)
    return case.theta_z_300 * rho_ratio


def material_properties(case: StackCase, temperatures_k: Any) -> dict[str, Any]:
    """Return temperature dependent magnetic and SOT properties."""

    temperatures = XP.atleast_1d(XP.asarray(temperatures_k, dtype=XP.float64))
    ms_ratio = saturation_ratio(
        temperatures,
        LOW_TEMP_MS_RATIO,
        PROPERTY_SATURATION_EXPONENT,
    )
    hk_ratio = saturation_ratio(
        temperatures,
        LOW_TEMP_HK_RATIO,
        PROPERTY_SATURATION_EXPONENT,
    )
    alpha_ratio = saturation_ratio(
        temperatures,
        LOW_TEMP_ALPHA_RATIO,
        PROPERTY_SATURATION_EXPONENT,
    )
    return {
        "theta_y": theta_y_temperature(case, temperatures),
        "theta_z": theta_z_temperature(case, temperatures),
        "ms_a_m": case.ms_300_a_m * ms_ratio,
        "alpha": case.alpha_300 * alpha_ratio,
        "effective_field_t": case.effective_field_300_t * hk_ratio,
    }


def magnon_occupation(frequency_ghz: Any, temperature_k: Any) -> Any:
    """Return the Bose occupation of a magnon mode."""

    frequency = XP.asarray(frequency_ghz, dtype=XP.float64) * 1.0e9
    temperature = XP.asarray(temperature_k, dtype=XP.float64)
    exponent = H_J_S * frequency / (K_B_J_K * temperature)
    return 1.0 / XP.expm1(exponent)


def fmr_frequency_ghz(effective_field_t: Any) -> Any:
    """Return a perpendicular macrospin FMR frequency in GHz."""

    field = XP.asarray(effective_field_t, dtype=XP.float64)
    return GAMMA_GHZ_T * XP.abs(field)


def field_from_frequency_t(frequency_ghz: Any) -> Any:
    """Return the effective field corresponding to an FMR frequency."""

    frequency = XP.asarray(frequency_ghz, dtype=XP.float64)
    return frequency / GAMMA_GHZ_T


def sot_effective_field_t(
    theta_eff: Any,
    current_density_a_cm2: Any,
    ms_a_m: Any,
    ferromagnet_thickness_nm: float,
) -> Any:
    """Return the damping like SOT effective field in tesla."""

    current_a_m2 = XP.asarray(current_density_a_cm2, dtype=XP.float64) * 1.0e4
    thickness_m = ferromagnet_thickness_nm * 1.0e-9
    return HBAR_J_S * theta_eff * current_a_m2 / (
        2.0 * E_CHARGE_C * ms_a_m * thickness_m
    )


def threshold_current_a_cm2(
    theta_eff: Any,
    alpha: Any,
    ms_a_m: Any,
    ferromagnet_thickness_nm: float,
    effective_field_t: Any,
) -> Any:
    """Return the antidamping threshold current density."""

    theta_abs = XP.maximum(XP.abs(theta_eff), 1.0e-12)
    thickness_m = ferromagnet_thickness_nm * 1.0e-9
    current_a_m2 = (
        2.0
        * E_CHARGE_C
        * alpha
        * ms_a_m
        * thickness_m
        * XP.abs(effective_field_t)
        / (HBAR_J_S * theta_abs)
    )
    return current_a_m2 / 1.0e4


def response_spectrum(
    frequency_axis_ghz: Any,
    mode_frequency_ghz: float,
    alpha_eff: float,
    drive_field_t: float,
) -> Any:
    """Return a normalized driven magnon response spectrum."""

    frequency = XP.asarray(frequency_axis_ghz, dtype=XP.float64)
    linewidth_ghz = max(2.0 * abs(alpha_eff) * mode_frequency_ghz, 1.0e-6)
    response = drive_field_t**2 / (
        (frequency - mode_frequency_ghz) ** 2 + (0.5 * linewidth_ghz) ** 2
    )
    max_response = XP.max(response)
    return response / XP.maximum(max_response, 1.0e-30)


def switching_proxy(
    current_a_cm2: Any,
    threshold_a_cm2: float,
    width_a_cm2: float,
    field_free_enabled: bool,
) -> Any:
    """Return a compact anomalous Hall switching proxy."""

    current = XP.asarray(current_a_cm2, dtype=XP.float64)
    if not field_free_enabled:
        return XP.zeros_like(current)

    branch_forward = XP.tanh((current - threshold_a_cm2) / width_a_cm2)
    branch_reverse = XP.tanh((current + threshold_a_cm2) / width_a_cm2)
    return 0.5 * (branch_forward + branch_reverse)


def build_summary_table() -> pd.DataFrame:
    """Build a table of cryogenic magnon and threshold quantities."""

    rows = []
    temperatures = XP.asarray(TEMPERATURES_K, dtype=XP.float64)
    mode_freq = TARGET_MODE_FREQUENCY_GHZ
    mode_field = field_from_frequency_t(mode_freq)

    for case in STACK_CASES:
        properties = material_properties(case, temperatures)
        j_threshold = threshold_current_a_cm2(
            properties["theta_y"],
            properties["alpha"],
            properties["ms_a_m"],
            case.ferromagnet_thickness_nm,
            mode_field,
        )
        occupation = magnon_occupation(mode_freq, temperatures)
        fmr = fmr_frequency_ghz(properties["effective_field_t"])
        theta_z_ratio = XP.abs(
            properties["theta_z"] / XP.maximum(
                XP.abs(properties["theta_y"]),
                1.0e-12,
            )
        )

        for index, temperature in enumerate(to_numpy(temperatures)):
            rows.append(
                {
                    "stack": case.name,
                    "T_K": temperature,
                    "mode_f_GHz": mode_freq,
                    "nbar": to_numpy(occupation)[index],
                    "theta_y": to_numpy(properties["theta_y"])[index],
                    "abs_theta_z_over_theta_y": to_numpy(theta_z_ratio)[index],
                    "FMR_from_default_field_GHz": to_numpy(fmr)[index],
                    "J_antidamping_A_cm2": to_numpy(j_threshold)[index],
                }
            )

    return pd.DataFrame(rows)


def plot_occupation_curves() -> None:
    """Plot thermal magnon occupation versus frequency."""

    frequency = XP.asarray(FREQUENCY_RANGE_GHZ, dtype=XP.float64)

    plt.figure()
    for temperature in PLOT_TEMPERATURES_K:
        occupation = magnon_occupation(frequency, temperature)
        plt.semilogy(
            to_numpy(frequency),
            to_numpy(occupation),
            label=f"{temperature:g} K",
        )

    plt.axhline(1.0, linestyle=":", linewidth=1.0)
    plt.xlabel("Mode frequency (GHz)")
    plt.ylabel(r"Thermal occupation $\bar{n}$")
    plt.title("Magnon occupation in the cryogenic regime")
    plt.grid(True, which="both", alpha=0.3)
    plt.legend()
    plt.show()


def plot_quantum_window_heatmap() -> None:
    """Plot the thermal occupation heatmap for magnon modes."""

    frequencies = XP.asarray(FREQUENCY_GRID_GHZ, dtype=XP.float64)
    temperatures = XP.asarray(TEMPERATURE_GRID_K, dtype=XP.float64)
    occupation = magnon_occupation(
        frequencies[None, :],
        temperatures[:, None],
    )
    log_occupation = XP.log10(occupation + 1.0e-12)

    plt.figure()
    mesh = plt.pcolormesh(
        to_numpy(frequencies),
        to_numpy(temperatures),
        to_numpy(log_occupation),
        shading="auto",
    )
    plt.yscale("log")
    plt.xlabel("Mode frequency (GHz)")
    plt.ylabel("Temperature (K)")
    plt.title(r"Quantum window, shown as $\log_{10}(\bar{n})$")
    plt.colorbar(mesh, label=r"$\log_{10}(\bar{n})$")
    plt.show()


def plot_threshold_vs_frequency(temperature_k: float) -> None:
    """Plot SOT antidamping threshold versus magnon frequency."""

    frequencies = XP.asarray(FREQUENCY_RANGE_GHZ, dtype=XP.float64)
    fields = field_from_frequency_t(frequencies)
    temperature = XP.asarray([temperature_k], dtype=XP.float64)

    plt.figure()
    for case in STACK_CASES:
        properties = material_properties(case, temperature)
        j_threshold = threshold_current_a_cm2(
            properties["theta_y"][0],
            properties["alpha"][0],
            properties["ms_a_m"][0],
            case.ferromagnet_thickness_nm,
            fields,
        )
        plt.semilogy(
            to_numpy(frequencies),
            to_numpy(j_threshold),
            label=case.name,
        )

    plt.xlabel("Mode frequency (GHz)")
    plt.ylabel(r"Antidamping threshold $J_\mathrm{th}$ (A cm$^{-2}$)")
    plt.title(f"SOT magnon threshold at {temperature_k:g} K")
    plt.grid(True, which="both", alpha=0.3)
    plt.legend(fontsize=7)
    plt.show()


def plot_threshold_vs_temperature() -> None:
    """Plot threshold current versus temperature for one target mode."""

    temperatures = XP.asarray(TEMPERATURES_K, dtype=XP.float64)
    mode_field = field_from_frequency_t(TARGET_MODE_FREQUENCY_GHZ)

    plt.figure()
    for case in STACK_CASES:
        properties = material_properties(case, temperatures)
        j_threshold = threshold_current_a_cm2(
            properties["theta_y"],
            properties["alpha"],
            properties["ms_a_m"],
            case.ferromagnet_thickness_nm,
            mode_field,
        )
        plt.loglog(
            to_numpy(temperatures),
            to_numpy(j_threshold),
            marker="o",
            label=case.name,
        )

    plt.gca().invert_xaxis()
    plt.xlabel("Temperature (K)")
    plt.ylabel(r"Antidamping threshold $J_\mathrm{th}$ (A cm$^{-2}$)")
    plt.title(f"Threshold for a {TARGET_MODE_FREQUENCY_GHZ:g} GHz mode")
    plt.grid(True, which="both", alpha=0.3)
    plt.legend(fontsize=7)
    plt.show()


def plot_drive_ratio_map(case: StackCase, temperature_k: float) -> None:
    """Plot normalized SOT drive for a heterostructure case."""

    frequencies = XP.asarray(FREQUENCY_GRID_GHZ, dtype=XP.float64)
    currents = XP.asarray(CURRENT_GRID_A_CM2, dtype=XP.float64)
    fields = field_from_frequency_t(frequencies)
    temperature = XP.asarray([temperature_k], dtype=XP.float64)
    properties = material_properties(case, temperature)
    j_threshold = threshold_current_a_cm2(
        properties["theta_y"][0],
        properties["alpha"][0],
        properties["ms_a_m"][0],
        case.ferromagnet_thickness_nm,
        fields[None, :],
    )
    drive_ratio = currents[:, None] / j_threshold

    plt.figure()
    mesh = plt.pcolormesh(
        to_numpy(frequencies),
        to_numpy(currents),
        to_numpy(drive_ratio),
        shading="auto",
    )
    plt.contour(
        to_numpy(frequencies),
        to_numpy(currents),
        to_numpy(drive_ratio),
        levels=[1.0],
        linewidths=1.0,
    )
    plt.xlabel("Mode frequency (GHz)")
    plt.ylabel(r"Charge current density (A cm$^{-2}$)")
    plt.title(f"Normalized SOT drive for {case.name} at {temperature_k:g} K")
    plt.colorbar(mesh, label=r"$J / J_\mathrm{th}$")
    plt.show()


def plot_driven_response(temperature_k: float) -> None:
    """Plot normalized driven magnon spectra for the stack cases."""

    frequency_axis = XP.asarray(FREQUENCY_RANGE_GHZ, dtype=XP.float64)
    temperature = XP.asarray([temperature_k], dtype=XP.float64)
    mode_field = field_from_frequency_t(TARGET_MODE_FREQUENCY_GHZ)

    plt.figure()
    for case in STACK_CASES:
        properties = material_properties(case, temperature)
        b_drive = sot_effective_field_t(
            properties["theta_y"][0],
            DRIVE_CURRENT_A_CM2,
            properties["ms_a_m"][0],
            case.ferromagnet_thickness_nm,
        )
        j_threshold = threshold_current_a_cm2(
            properties["theta_y"][0],
            properties["alpha"][0],
            properties["ms_a_m"][0],
            case.ferromagnet_thickness_nm,
            mode_field,
        )
        alpha_eff = float(
            to_numpy(properties["alpha"])[0]
            * (1.0 - DRIVE_CURRENT_A_CM2 / to_numpy(j_threshold))
        )
        response = response_spectrum(
            frequency_axis,
            TARGET_MODE_FREQUENCY_GHZ,
            alpha_eff,
            float(to_numpy(b_drive)),
        )
        plt.plot(
            to_numpy(frequency_axis),
            to_numpy(response),
            label=case.name,
        )

    plt.xlabel("Frequency (GHz)")
    plt.ylabel("Normalized driven magnon response")
    plt.title(
        f"Driven spectra at {temperature_k:g} K "
        f"and J = {DRIVE_CURRENT_A_CM2:.1e} A cm$^{{-2}}$"
    )
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=7)
    plt.show()


def plot_switching_proxy() -> None:
    """Plot a compact field free switching proxy."""

    currents = XP.linspace(-3.5e7, 3.5e7, 800)
    ni4w_curve = switching_proxy(
        currents,
        NI4W_CO_REPORTED_SWITCHING_A_CM2,
        2.5e6,
        field_free_enabled=True,
    )
    pt_zero_field_curve = switching_proxy(
        currents,
        PT_CO_REPORTED_SWITCHING_A_CM2,
        3.0e6,
        field_free_enabled=False,
    )
    pt_assisted_curve = switching_proxy(
        currents,
        PT_CO_REPORTED_SWITCHING_A_CM2,
        3.0e6,
        field_free_enabled=True,
    )

    plt.figure()
    plt.plot(
        to_numpy(currents),
        to_numpy(ni4w_curve),
        label="Ni4W / Co, field free proxy",
    )
    plt.plot(
        to_numpy(currents),
        to_numpy(pt_zero_field_curve),
        label="Pt / Co, zero field proxy",
    )
    plt.plot(
        to_numpy(currents),
        to_numpy(pt_assisted_curve),
        label="Pt / Co, assisted field proxy",
    )
    plt.xlabel(r"Current density (A cm$^{-2}$)")
    plt.ylabel("Normalized anomalous Hall response")
    plt.title("Switching proxy anchored to reported room temperature devices")
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=7)
    plt.show()


def main() -> pd.DataFrame:
    """Run the magnonic model and render all plots."""

    backend = "CuPy GPU" if USING_CUPY else "NumPy CPU"
    print(f"Backend: {backend}")
    print(f"Gyromagnetic ratio: {GAMMA_GHZ_T:.3f} GHz/T")

    table = build_summary_table()
    display_columns = [
        "stack",
        "T_K",
        "mode_f_GHz",
        "nbar",
        "theta_y",
        "abs_theta_z_over_theta_y",
        "J_antidamping_A_cm2",
    ]
    display(table[display_columns])

    plot_occupation_curves()
    plot_quantum_window_heatmap()
    plot_threshold_vs_frequency(DRIVE_MAP_TEMPERATURE_K)
    plot_threshold_vs_temperature()
    plot_drive_ratio_map(STACK_CASES[2], DRIVE_MAP_TEMPERATURE_K)
    plot_driven_response(MAGNON_RESPONSE_TEMPERATURE_K)
    plot_switching_proxy()

    return table


if __name__ == "__main__":
    results = main()
```
