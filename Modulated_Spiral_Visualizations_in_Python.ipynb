The script below plots a Gaussian modulated spiral and a Blackman-Harris modulated spiral. If plotted or modified carefully, it can be helpful to visualize a wavefunction on a 3D plot and converted into Blender Python script.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Onri prefers using rcParams to increase the quality of all plots to be higher than the default settings.
# This only needs to be done once at beginning of every Jupyter Notebook.
plt.rcParams['figure.dpi'] = 200

# Parameters
A = 1          # Scaling factor
sigma = 10     # Gaussian width
theta_max = 10 * np.pi  # Maximum angle (5 full turns)
num_points = 1000       # Number of points
z_max = 5               # Maximum height for the spiral

# Generate theta values
theta = np.linspace(0, theta_max, num_points)
x = np.linspace(0, z_max, num_points)  # X-axis variation (formerly Z)

# Compute radius with Gaussian modulation
r = A * np.exp(-theta**2 / (2 * sigma**2)) * theta

# Convert to Cartesian coordinates
z = r * np.cos(theta)  # Rotated to z-axis
y = r * np.sin(theta)  # Rotated to y-axis

# Create 3D plot
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the rotated 3D spiral
ax.plot(x, y, z, linewidth=2, color='b')

# Labels and view adjustment
ax.set_title("Rotated Gaussian-Modulated 3D Spiral")
ax.set_xlabel("X (Height)")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.grid(False)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 1          # Scaling factor
sigma = 10     # Gaussian width
theta_max = 10 * np.pi  # Maximum angle (5 full turns)
num_points = 1000       # Number of points

# Generate theta values
theta = np.linspace(0, theta_max, num_points)

# Compute radius with Gaussian modulation
r = A * np.exp(-theta**2 / (2 * sigma**2)) * theta

# Convert to Cartesian coordinates
x = r * np.cos(theta)
y = r * np.sin(theta)

# Plot the spiral
plt.figure(figsize=(8, 8))
plt.plot(x, y, linewidth=2, color='b')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axis('equal')
plt.title("Gaussian-Modulated Spiral")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal.windows import blackmanharris

# Parameters
num_points = 1000       # Number of points
theta_max = 10 * np.pi  # Maximum angle (5 full turns)
x_max = 5               # Maximum height for the spiral

# Generate theta values
theta = np.linspace(0, theta_max, num_points)
x = np.linspace(0, x_max, num_points)  # Linearly increasing in the x-direction (formerly z)

# Compute radius
r = theta

# Apply Blackman-Harris window
window = blackmanharris(num_points)
r_modulated = r * window

# Convert to Cartesian coordinates (rotated)
z = r_modulated * np.cos(theta)  # Moved to z-axis
y = r_modulated * np.sin(theta)  # Stays on y-axis

# Create 3D plot
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the rotated 3D spiral
ax.plot(x, y, z, linewidth=2, color='b')

# Labels and view adjustment
ax.set_title("Rotated Blackman-Harris Modulated 3D Spiral")
ax.set_xlabel("X (Height)")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.grid(False)

# Show the plot
plt.show()
```
