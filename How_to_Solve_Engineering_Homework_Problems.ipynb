```python
import plotly.graph_objects as go

# Define nodes

nodes = [
    "Highlight Key Terms",
    "Record Linguistic Definitions",
    "Locate Required Equations",
    "Convert Units & Formulas",
    "List or Tabulate Assumed/Ignored Terms",
    "Script into LaTeX/Markdown",
    "Import the LaTeX Script in Google Colab",
    "Solve Problems Sequentially with Python",
    "Double-check with Textbooks",
    "Validate with Research Literature"
]

# Define links

sources = [0, 0, 2, 2, 4, 5, 6, 7, 7]

targets = [1, 2, 3, 4, 5, 6, 7, 8, 9]

values = [1, 1, 1, 1, 1, 1, 1, 1, 1]

# Create Sankey diagram

fig = go.Figure(go.Sankey(
    node=dict(
        pad=20,
        thickness=20,
        line=dict(color='black', width=0.5),
        label=nodes
    ),
    link=dict(
        source=sources,
        target=targets,
        value=values
        )
))

fig.update_layout(
    title=dict(text="How to Solve Engineering Homework Problems by O.J.B. - Sankey Diagram", font=dict(size=20)),
    font=dict(size=12),
    autosize=False,
    width=1000,
    height=600
)

fig.show(config={'displayModeBar': True, 'staticPlot': False, 'toImageButtonOptions': {'format': 'png', 'scale': 3}})
```
