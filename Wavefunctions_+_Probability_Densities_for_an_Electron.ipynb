```python
# Plotted wavefunctions and probability densities for an electron in a one-dimensional box of length 1 nm
import numpy as np
import matplotlib.pyplot as plt

# Constants
hbar = 1.0545718e-34  # Reduced Planck constant (Joule-second)
m_e = 9.10938356e-31  # Electron mass (kg)
L = 1e-9  # Length of the box (1 nm in meters)

# Quantum number (n=1,2,3 for first three states)
n_values = [1, 2, 3]

# Define the wavefunction for a particle in a box
def psi_n(x, n):
    return np.sqrt(2 / L) * np.sin(n * np.pi * x / L)

# Define the energy for a particle in a box
def energy_n(n):
    return (n**2 * np.pi**2 * hbar**2) / (2 * m_e * L**2)

# Generate x values
x_values = np.linspace(0, L, 1000)

# Plotting with a DPI of 200
plt.figure(figsize=(12, 8), dpi=300)

for n in n_values:
    psi_values = psi_n(x_values, n)
    probability_density = np.abs(psi_values)**2

    # Plot wavefunction
    plt.subplot(2, 1, 1)
    plt.plot(x_values, psi_values, label=f'n={n}')
    plt.title('Wavefunctions of an Electron in a Box')
    plt.xlabel('Position (m)')
    plt.ylabel('Wavefunction ')
    plt.legend()

    # Plot probability density
    plt.subplot(2, 1, 2)
    plt.plot(x_values, probability_density, label=f'n={n}')
    plt.title('Probability Densities of an Electron in a Box')
    plt.xlabel('Position (m)')
    plt.ylabel('Probability Density ||^2')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Replotting with specific color schemes: red for n=1, blue for n=2, and green for n=3. Box still has a length of 1 nm

colors = ['red', 'blue', 'green']

plt.figure(figsize=(12, 8), dpi=250)

for i, n in enumerate(n_values):
    psi_values = psi_n(x_values, n)
    probability_density = np.abs(psi_values)**2
    color = colors[i]

    # Plot wavefunction
    plt.subplot(2, 1, 1)
    plt.plot(x_values, psi_values, label=f'n={n}', color=color)
    plt.title('Wavefunctions of an Electron in a Box')
    plt.xlabel('Position (m)')
    plt.ylabel('Wavefunction ')
    plt.legend()

    # Plot probability density
    plt.subplot(2, 1, 2)
    plt.plot(x_values, probability_density, label=f'n={n}', color=color)
    plt.title('Probability Densities of an Electron in a Box')
    plt.xlabel('Position (m)')
    plt.ylabel('Probability Density ||^2')
    plt.legend()

plt.tight_layout()
plt.show()
```
