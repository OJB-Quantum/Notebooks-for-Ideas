```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.quantum_info import Operator
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
# Onri prefers using rcParams to increase the quality of all plots to be higher than the default settings.
# This only needs to be done once at beginning of every Jupyter Notebook.
import matplotlib as plt
plt.rcParams['figure.dpi'] = 250
```

```python
# @title Build and render the MPL quantum circuit layout
"""Create and render a custom two-qubit quantum circuit."""

import numpy as np
from IPython.display import display

qc = QuantumCircuit(2)

qc.h(0)                  # Hadamard on q0
qc.cx(0, 1)              # CNOT with q0 as control and q1 as target
qc.ry(np.pi / 4, 1)      # Y-rotation of pi/4 on q1
qc.z(1)                  # Phase flip (Z gate) on q1

circuit_figure = qc.draw(
    output="mpl",
    style="iqp",
    fold=-1,
    idle_wires=True,
)

display(circuit_figure)
display(Operator(qc).draw("latex"))
```
