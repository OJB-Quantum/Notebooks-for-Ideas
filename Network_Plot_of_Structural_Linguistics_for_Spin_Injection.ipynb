```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Defining the main terms from the document
terms = ["Spin Injection", "Electron Spin", "Condensed Matter Physics",
         "Spin-dependent Transport", "Fermi Level", "Spin Polarization",
         "Spin Accumulation", "Metals", "Semiconductors", "Giant Magnetoresistance",
         "Spintronics", "Interface Effects", "Magnetic Fields"]

# Creating a network graph
G = nx.Graph()

# Adding nodes
for term in terms:
    G.add_node(term)

# Adding edges between related terms
edges = [("Spin Injection", "Electron Spin"), ("Spin Injection", "Condensed Matter Physics"),
         ("Spin Injection", "Spin-dependent Transport"), ("Spin Injection", "Fermi Level"),
         ("Spin Injection", "Spin Polarization"), ("Spin Injection", "Spin Accumulation"),
         ("Spin Injection", "Metals"), ("Spin Injection", "Semiconductors"),
         ("Spin Injection", "Giant Magnetoresistance"), ("Spin Injection", "Spintronics"),
         ("Spin Injection", "Interface Effects"), ("Spin Injection", "Magnetic Fields"),
         ("Electron Spin", "Condensed Matter Physics"), ("Electron Spin", "Spin-dependent Transport"),
         ("Spin-dependent Transport", "Fermi Level"), ("Spin-dependent Transport", "Spin Polarization"),
         ("Spin-dependent Transport", "Metals"), ("Spin-dependent Transport", "Semiconductors"),
         ("Fermi Level", "Metals"), ("Fermi Level", "Semiconductors"),
         ("Spin Polarization", "Spin Accumulation"), ("Spin Polarization", "Giant Magnetoresistance"),
         ("Spin Accumulation", "Metals"), ("Spin Accumulation", "Semiconductors"),
         ("Metals", "Giant Magnetoresistance"), ("Semiconductors", "Giant Magnetoresistance"),
         ("Giant Magnetoresistance", "Spintronics"), ("Spintronics", "Interface Effects"),
         ("Interface Effects", "Magnetic Fields"), ("Magnetic Fields", "Electron Spin")]

# Adding edges to the graph
G.add_edges_from(edges)

# Drawing the graph
plt.figure(figsize=(12, 12), dpi=300)
pos = nx.spring_layout(G, k=0.15, iterations=20)
nx.draw(G, pos, with_labels=True, node_color="skyblue", node_size=3000, edge_color="gray", linewidths=1, font_size=12)
plt.title("Network Plot of Main Terminology in 'Spin Injection'", size=20)
plt.show()
```
