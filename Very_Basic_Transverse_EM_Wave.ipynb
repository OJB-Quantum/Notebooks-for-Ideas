```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters for the wave
period = 6 * np.pi
t = np.linspace(0, period, 500)

# Create the sinusoidal waves
x_wave = np.sin(t)
y_wave = np.sin(t + np.pi / 2)  # Orthogonal wave (90-degree phase shift)
z = t

# Create a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the waves
ax.plot(x_wave, z, np.zeros_like(z), label='Electric Field', linewidth=2)
ax.plot(np.zeros_like(z), z, y_wave, label='Magnetic Field', linewidth=2)

# Remove grid lines and labels
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define parameters
period = 6 * np.pi
t = np.linspace(0, period, 500)

# Define orthogonal sinusoidal waves
x_wave = np.sin(t)  # Electric field
y_wave = np.sin(t + np.pi/2)  # Magnetic field

# Plot the waves
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(t, x_wave, np.zeros_like(t), label="Electric Field")
ax.plot(t, np.zeros_like(t), y_wave, label="Magnetic Field")

# Remove grid lines, labels, and title
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

plt.show()
```
