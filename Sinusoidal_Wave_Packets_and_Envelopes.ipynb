Below are some wave packets and their envelopes for visualizing modulated sine waves.

```python
import numpy as np
import matplotlib.pyplot as plt

# Onri prefers using rcParams to increase the quality of all plots to be higher than the default settings.
# This only needs to be done once at beginning of every Jupyter Notebook.
plt.rcParams['figure.dpi'] = 200

# Define parameters
x = np.linspace(-50, 50, 1000)  # x-axis range
wavelength = 10 * np.pi         # Period of the sinusoidal wave
k = 2 * np.pi / wavelength      # Wave number
sigma = 20                      # Standard deviation of the envelope

# Generate sinusoidal wave packet
wave_packet = np.sin(k * x) * np.exp(-x**2 / (2 * sigma**2))

# Generate envelope
envelope = np.exp(-x**2 / (2 * sigma**2))

# Plot wave packet and envelope
plt.figure(figsize=(10, 5))
plt.plot(x, wave_packet, label="Wave Packet", color='b')
plt.plot(x, envelope, '--', label="Envelope", color='r')
plt.plot(x, -envelope, '--', color='r')  # Negative envelope
plt.xlabel("x")
plt.ylabel("Amplitude")
plt.title("Sinusoidal Wave Packet with Envelope")
plt.legend()
# plt.grid()
plt.show()
```
